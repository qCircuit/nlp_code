```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier(["Today is a nice weather", "Сегодня хорошая погода"])
```

```python
classifier(["I feel bad", "That is a terrible problem while choosing an appartment"])
```

```python
classifier(["Я плохо себя чувствую", "Это ужасная проблема при выборе квартиры"])
```

```python
#@title zero shot classification
classifier = pipeline("zero-shot-classification")
```

```python
text = """That information then provides a rough picture of cryptocurrency activity, although as crypto wallets can be set up anonymously and without the owner specifying location, it is not necessarily precise. """
classifier(
    text,
    candidate_labels=["education", "finance", "politics", "money"]
)
```

```python
#@title text generation
generator = pipeline("text-generation")
```

```python
text = """if you have only paper and rocks you can"""

generator = pipeline("text-generation", model="distilgpt2")
generated_text = generator(
    text,
    num_return_sequences=3,
    max_length = 20
)

for text in generated_text:
    print(text)
```

```python
#@title mask filling

unmasker = pipeline("fill-mask")
```

```python
text = """Maria Sharapova is very good at playing <mask> games"""
unmasker(text, top_k=2)
```

```python
text = """На олимпийских <mask> соревнуются команды из разных стран"""
unmasker(text, top_k=2)
```

```python
#@title named entity recognition
ner = pipeline("ner", grouped_entities=True)
# grouped_entities: tell the pipeline to regroup together the parts of the sentence that correspond to the same entity
```

```python
text = """In 1895, at the age of sixteen, Einstein sat the entrance examination for the Federal polytechnic school (later the Eidgenössische Technische Hochschule, ETH) in Zürich, Switzerland, where he failed to reach the required standard in the general part of the test, but performed with distinction in physics and mathematics."""

ner(text)
```

```python
#@title question-answering
question_answerer = pipeline("question-answering")
```

```python
question_answerer(
    question="Who let the dogs out?",
    context="After having a dinner Mr Jones let all his dogs out for a walk"
)
```

```python
question_answerer(
    question="Кто выпустил собак?",
    context="После сытного обеда господин Плюшкин выпустил своих собак на прогулку"
)
```

```python
#@title summarization
summarizer = pipeline("summarization")
```

```python
text="""Assassin's Creed Mirage is released this week - and reviews have called it a return to the series' roots. But there's also been one major change.
For this edition, set in 9th Century Baghdad, developers Ubisoft have recorded the game's performances in Arabic.
Previous games in the long-running stealth franchise have taken place in the Middle East, Europe and Africa, but the heroes have always been voiced by English-speaking actors by default.
Players can still choose to play the game with English dialogue, but its makers hope they'll choose the more "authentic" Arabic."""

summarizer(text)
```

```python
text="""Двукратный финалист турниров «Большого шлема», шестая ракетка мира, греческий теннисист Стефанос Циципас не сумел пробиться в 1/8 финала турнира серии «Мастерс» в Шанхае (Китай), призовой фонд которого составил $ 8 800 000.
В матче третьего круга Циципас (4) со счётом 4:6, 6:3, 5:7 уступил 34-й ракетке мира, французу Уго Умберу (32). Встреча теннисистов продлилась 2 часа и 11 минут. Циципас семь раз подал навылет, допустил одну двойную ошибку и реализовал два брейк-пойнта из восьми. На счету Умбера восемь эйсов, одна двойная ошибка и три реализованных брейк-пойнта из шести заработанных.
За выход в четвертьфинал турнира в Шанхае Умбер сразится с 51-й ракеткой мира, американцем Джеффри-Джоном Вулфом."""

summarizer(text)
```
